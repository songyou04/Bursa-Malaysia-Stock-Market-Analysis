```python
import pandas as pd

#Load the dataset
file_data_path = 'MY Stock Market 2000-01-03 .csv'

# Read data file
with open(file_data_path, 'r') as f:
    data_content = f.readlines()  

# Define missing values
missing_values = ['n/a', 'na', '--', '?']

# Read data and replace missing values with NaN
df = pd.read_csv(file_data_path, na_values=missing_values)

# Display the first few rows of the dataset
print("Initial Dataset:")
display(df.head())
```

```python
# Explore the dataset by checking the number of rows and columns
print(f"Number of rows: {df.shape[0]}")
print(f"Number of columns: {df.shape[1]}")

# Display the range of values for key numerical columns
print("\nRange of Values for Key Numerical Columns:")
for col in ['Open', 'High', 'Low', 'Close']:
    if col in df.columns:
        min_val = df[col].min()
        max_val = df[col].max()
        value_range = max_val - min_val  # Calculate the range (max - min)
        print(f"{col}: Min = {min_val}, Max = {max_val}, Range = {value_range}")

# Check for unique values in categorical columns
print("\nUnique Values in Each Column:")
print(df.nunique())
```

```python
# Handle missing, incorrect, and invalid data
# Check for missing values before dropping
print("\nMissing Values Before Drop:")
print(df.isnull().sum())

# Drop rows with missing values
df_cleaned = df.dropna().copy()  
```

```python
# Check for missing values after dropping
print("\nMissing Values After Drop:")
print(df_cleaned.isnull().sum())

# Check for invalid data (negative values in numerical columns)
print("\nChecking for Invalid Data (e.g., Negative Prices):")
negative_values = df_cleaned[df_cleaned['Close'] < 0]  # Replace 'Close' with the correct column name
print(negative_values)

# Handle invalid data (replace negative values with NaN)
df_cleaned.loc[df_cleaned['Close'] < 0, 'Close'] = None  # Replace 'Close' with the appropriate column name
```

```python
# Parse the 'Date' column into datetime format
# parsing without specifying the format to handle mixed formats
df_cleaned['Date'] = pd.to_datetime(df_cleaned['Date'], errors='coerce')

# Drop rows with invalid date parsing
df_cleaned = df_cleaned.dropna(subset=['Date'])

# Format the date column to day/month/year format
df_cleaned['Date'] = df_cleaned['Date'].apply(lambda x: x.strftime('%d/%m/%Y'))

# Verify the conversion
print("\nDate Column Data Type After Parsing:")
print(df_cleaned['Date'].head())

# Create additional columns (Daily_Return)
df_cleaned['Daily_Return'] = df_cleaned['Close'].pct_change() # Daily return = (Close Price Today - Close Price Yesterday) / Close Price Yesterday

# Display the updated DataFrame
print("\nCleaned Dataset with Additional Columns:")
display(df_cleaned.head())
```
